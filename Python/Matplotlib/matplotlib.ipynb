## Visualization with NumPy and Matplotlib

```python
import matplotlib.pyplot as plt
```

```python
x = np.arange(6)
y = x + 1
print(x)
print(y)
```

```python
plt.plot(x,y, 'o--')
plt.show()
```

```python
plt.plot(x,y,'o--')
plt.plot(x, -y, 'o-')
```

```python
plt.title('y=x and y=-x')
```

```python
N = 16
x = np.linspace(0,15,N)
x
```

```python
y = x
plt.plot(x,y,'o--')
plt.axis('off')
```

```python
y = np.logspace(0.1,2,N)
plt.plot(x,y,'o--')
```

```python
y = np.geomspace(0.1, 2000, N)
plt.plot(x,y,'o--')
```

# Revisiting Matplotlib Visualizations

## Single-Line Plots

```python
x = [4,5,4,1,6,7]
plt.plot(x)
```

```python
x = np.arange(25)
plt.plot(x)
```

```python
plt.plot(x, [y ** 3 + 1 for y in x])
```

```python
plt.plot(x, x ** 3 + 1)
```

## Multiline Plots

```python
x = np.arange(7)
plt.plot(x, -x**2)
plt.plot(x, -x**3)
plt.plot(x, -2*x)
plt.plot(x, -2**x)
```

```python
plt.plot(x, -x**2, x, -x**3, x, -2*x, x, -2**x)
```

```python
x = np.array([[3,2,5,6],[7,4,1,5]])
plt.plot(x)
```

```python
plt.plot(np.random.randn(2,10))
```

## Grid, Axes, and Labels

```python
x = np.arange(3)
plt.plot(x, -x**2, x, -x**3, x, -2*x, x, -2**x)
plt.grid(True)
print(plt.axis())
```

```python
x = np.arange(3)
plt.plot(x, -x**2, x, -x**3, x, -2*x, x, -2**x)
plt.grid(True)
plt.axis([0,2,-8,0]) # sets the values of the axes(limits)
print(plt.axis())
```
