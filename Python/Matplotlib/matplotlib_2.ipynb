```python
n = 3
data = np.linspace(0, 2*n*np.pi, 300)

def sinusoidal(sty):
    plt.style.use(sty)
    fig, ax = plt.subplots()
    ax.plot(data, np.sin(data), label = 'Sine')
    ax.plot(data, np.cos(data), label = 'Cosine')
    ax.legend()
    
sinusoidal('default')
plt.show()
```

```python
sinusoidal('ggplot')
plt.show()
```

```python
sinusoidal('seaborn')
plt.show()
```

## Layouts

```python
plt.style.use('default')
fig, axs = plt.subplots(2,2 , constrained_layout = True)
plt.show()
```

- Use gridspec to create subplots as follows : https://zephyrus1111.tistory.com/226

```python
import matplotlib.gridspec as gridspec

fig = plt.figure(constrained_layout = True)
specs = gridspec.GridSpec(ncols=2, nrows=2, figure=fig)
ax1 = fig.add_subplot(specs[0,0])
ax2 = fig.add_subplot(specs[0,1])
ax3 = fig.add_subplot(specs[1,0])
ax4 = fig.add_subplot(specs[1,1])
plt.show()
```

```python
fig = plt.figure(constrained_layout = True)
gs = fig.add_gridspec(3,3)
ax1 = fig.add_subplot(gs[0,:])
ax1.set_title('gs[0,:]')
ax2 = fig.add_subplot(gs[1, :])
ax2.set_title('gs[1,:]')
ax3 = fig.add_subplot(gs[2, :])
ax3.set_title('gs[2, :]')
plt.show()
```

```python
fig = plt.figure(constrained_layout = True)
gs = fig.add_gridspec(3,3)
ax1 = fig.add_subplot(gs[:,0])
ax1.set_title('gs[:,0]')
ax2 = fig.add_subplot(gs[:, 1])
ax2.set_title('gs[:,1]')
ax3 = fig.add_subplot(gs[:, 2])
ax3.set_title('gs[:,2]')
plt.show()
```

```python
fig = plt.figure(constrained_layout = True)
gs = fig.add_gridspec(3,3)
ax1 = fig.add_subplot(gs[0,:])
ax1.set_title('gs[0,:]')
ax2 = fig.add_subplot(gs[1, :-1])
ax2.set_title('gs[1,:-1]')
ax3 = fig.add_subplot(gs[1:, -1])
ax3.set_title('gs[1:-1]')
ax4 = fig.add_subplot(gs[-1,0])
ax4.set_title('gs[-1,0]')
ax5 = fig.add_subplot(gs[-1,-2])
ax5.set_title('gs[-1,-2]')
plt.show()
```

# Histograms, Contours, and Stream Plots

# Histograms

```python
x = [1, 3, 5, 1, 2, 4, 4, 2, 5, 4, 3, 1, 2]
n_bins = 5
plt.hist(x, bins=n_bins)
plt.show()
```

```python
np.random.seed(31425)
n_points = 1000
n_bins = 15
x = np.random.randn(n_points)
plt.hist(x, bins = n_bins)
plt.show()
```

```python
y = np.random.randn(n_points)
plt.hist2d(x, y, bins=50)
plt.show()
```

```python
plt.hist(x, 20, density=True, histtype='stepfilled', facecolor = 'g', alpha = 0.5)
plt.show()
```

```python
plt.hist(x,20, density=True, histtype='step')
plt.show()
```

## Countours

```python
x = np.arange(-3, 3, 0.005)
y = np.arange(-3, 3, 0.005)
x,y = np.meshgrid(x,y)
z = x**2 + y**2
out = plt.contour(x,y,z)
plt.clabel(out, inline = True, fontsize = 10)
plt.show()
```

- Add Color bar to the ouput as follows

```python
out = plt.contour(x,y,z)
plt.clabel(out, inline = True, fontsize = 10)
plt.colorbar(out)
plt.show()
```

```python
plt.contourf(x,y,z, hatches=['-','/','\\','//'], cmap = 'cool', alpha = 0.75)
plt.show()
```

## Visualizing Vectors with Stream Plots

```python
Y, X = np.mgrid[-5:5:200j, -5:5:300j]
U = X**2 + Y**2
V = X + Y
plt.streamplot(X, Y , U, V)
```

```python
plt.streamplot(X,Y,U,V, density=[0.5, 0.75])
```

```python
plt.streamplot(X,Y,U,V, color=V, linewidth=1, cmap='cool')
```

```python
plt.streamplot(X, Y, U, V, density=0.6, color='k', linewidth=X)
```

```python
X = np.arange(-5,5,0.5)
Y = np.arange(-10, 10, 1)
U, V = np.meshgrid(X, Y)
plt.quiver(X, Y, U, V)
plt.show()
```

## Polar Charts

```python
N = 20
theta = np.linspace(0.0, 2 * np.pi, N)
r = 10 * np.random.rand(N)
plt.subplot(projection = 'polar')
plt.bar(theta, r, bottom=0.0, color = ['r','g','b'], alpha = 0.2)
```

```python
r = np.arange(0, 5, 0.2)
theta = 2 * np.pi * r
plt.subplot(projection='polar')
plt.plot(theta, r)
```

```python
r = np.arange(0, 5, 0.01)
theta = 2 * np.pi * r
plt.subplot(projection='polar')
plt.plot(theta, r)
```

```python
N = 150
r = np.random.rand(N)
theta = 2 * np.pi * np.random.rand(N)
size = r * 100

plt.subplot(projection = 'polar')
plt.scatter(theta, r, c = theta, s = size, cmap = 'hsv', alpha = 0.5)
```

```python
fig = plt.figure()
ax = fig.add_subplot(projection = 'polar')
c = ax.scatter(theta, r, c = theta, s = size, cmap = 'hsv', alpha=0.5)
ax.set_thetamin(0)
ax.set_thetamax(90)
```
