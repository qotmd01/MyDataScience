## Visualization with NumPy and Matplotlib

````python
import matplotlib.pyplot as plt
````

````python
x = np.arange(6)
y = x + 1
print(x)
print(y)
````

````python
plt.plot(x,y, 'o--')
plt.show()
````

````python
plt.plot(x,y,'o--')
plt.plot(x, -y, 'o-')
````

````python
plt.title('y=x and y=-x')
````

````python
N = 16
x = np.linspace(0,15,N)
x
````

````python
y = x
plt.plot(x,y,'o--')
plt.axis('off')
````

````python
y = np.logspace(0.1,2,N)
plt.plot(x,y,'o--')
````

````python
y = np.geomspace(0.1, 2000, N)
plt.plot(x,y,'o--')
````

## Grid, Axes, and Labels

````python
x = np.arange(3)
plt.plot(x, -x**2, x, -x**3, x, -2*x, x, -2**x)
plt.grid(True)
print(plt.axis())
````

````python
x = np.arange(3)
plt.plot(x, -x**2, x, -x**3, x, -2*x, x, -2**x)
plt.grid(True)
plt.axis([0,2,-8,0]) # sets the values of the axes(limits)
print(plt.axis())
````

````python
x = np.arange(3)
plt.plot(x, -x**2, x, -x**3, x, -2*x, x, -2**x)
plt.grid(True)
plt.xlabel('x = np.arange(3)')
plt.xlim([0,2])
plt.ylabel('y = f(x)')
plt.ylim([-8,0])
plt.title('Simple Plot Demo')
plt.show()
````

````python
x = np.arange(3)
plt.plot(x, -x**2, label = '-x**2')
plt.plot(x, -x**3, label = '-x**3')
plt.plot(x, -2*x, label = '-2*x')
plt.plot(x, -2**x, label = '-2**x')
plt.legend()
plt.grid(True)
plt.xlabel('x = np.arange(3)')
plt.xlim([0,2])
plt.ylabel('y = f(x)')
plt.ylim([-8,0])
plt.title('Simple Plot Demo')
plt.show()
````

````python
x = np.arange(3)
plt.plot(x, -x**2, x, -x**3, x, -2*x, x, -2**x)
plt.legend([' -x**2',  '-x**3', '-2*x', '-2**x'], loc = 'lower center')
plt.grid(True)
plt.xlabel('x = np.arange(3)')
plt.xlim([0,2])
plt.ylabel('y = f(x)')
plt.ylim([-8,0])
plt.title('Simple Plot Demo')
plt.show()
````

## Object-Oriented Plotting

````python
x = np.arange(3)
fig, ax = plt.subplots()
ax.plot(x, -x**2, label = '-x**2')
ax.plot(x, -x**3, label = '-x**3')
ax.plot(x , -2*x, label = '-2*x')
ax.plot(x, -2**x, label = '-2**x')
ax.set_xlabel('x = np.arange(3)')
ax.set_ylabel('y = f(x)')
ax.set_title('Simple Plot Demo')
ax.legend()
ax.grid(True)
ax.text(0.25, -5, 'Simple Plot Demo')
plt.show()
````

## Subplot

- Subplotting : Show multiple separate graphs in the same output  
- Subplots have their own titles, own labels, and other specifications

```python
plt.subplot(grid_size_row, grid_size_column, position)
```

````python
x = np.arange(3)
fig,ax = plt.subplots(2,2)
ax[0][0].plot(x, -x**2)
ax[0][1].plot(x, -x**3)
ax[1][0].plot(x, -2*x)

plt.subplots_adjust(wspace=0.3, hspace=0.3)
````
